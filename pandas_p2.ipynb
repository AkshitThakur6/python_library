```python
#add a column total= python + machine learning
df['total']=df['Python']+df['Machine Learning']
df
```

```python
def grade(Marks):
    p=(Marks/200)*100
    if (p>75):
        return 'A'
    elif (p>60 and p<75):
        return 'b'
    elif (p>35 and p<60):
        return 'c'
    else:
        return 'f'
```

```python
#apply grade to function
df['grade']=df["total"].apply(grade)# apply is use to apply grade in df 
df
```
