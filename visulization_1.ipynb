```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
area=np.array([600 ,800, 1000, 1500, 2000])
price=np.array([4,5,9,14,22])
print(area)
print(price)
```

```python
#scatter plot by using above data
plt.scatter(area,price,c='green',marker='*')
plt.xlabel('Area in sqft',c='blue')
plt.ylabel('Price in lakh',c='blue')
# plt.title(label='Area v/s price')
plt.title("Area v/s Price")
plt.show()
```

```python
#line plot
plt.plot(area,price,c='red')
plt.scatter(area,price)
plt.xlabel('area of sqft',c='blue')
plt.ylabel('price in lakh',c='blue')
plt.title('area vs price',c='blue')
plt.show()
```

```python
#visulizating data by seaborn
data={
    'Area':[600, 800 ,1000 ,1500, 2000],
    'price':[4 ,5, 9, 14, 22]
}
df=pd.DataFrame(data)
df
```

```python
X=df['Area']
Y=df['price']
sns.scatterplot(data=df,x=X,y=Y)
sns.lineplot(x=X,y=Y)
plt.title('Area v/s price')
plt.show()
```
