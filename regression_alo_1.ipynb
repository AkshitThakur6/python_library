<h1>Regression:</h1>
It is method for understanding relationship between dependant and indepedant variable.<br>It is used on continous dataset <br>
regression 2 types
1.simple--
 simple linear Regression<br>
 simple non-linear<br>
2. Multiple regression--><br>
  multiple linear<br>
  multiple non linear regression

  if one of feature is mapped with target then it is <b>simple regression</b><br>
  if this selected feature and target have linear relationship then it is <b>simple linear regression</b><br>
  if it is show non-linearity b/w feature and target then it is <b>simple non-linear regression</b><br>
if more than one feature are mapped for target then it will be <b> multiple regression</b>




```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#y=mx+c    m----slope , c-----intercept, x-----i/p
x=np.array([1,2,3,4])
slope=0.5
intercept=4
y=slope*x+intercept
y
```

```python
plt.figure()
plt.plot(x,y)
plt.scatter(x,y)
plt.show() # this show linear regression beacuse x value will be increase and y also increase with x
```

```python
area=np.array([700,750,800,900,1000])
price=np.array([20,30,33,45,80])
plt.figure()
plt.scatter(area,price)
plt.plot(area,price)
plt.show()
```

```python
#predicting price
ypred=area*slope+intercept
```

```python
area_m=np.mean(area)
price_m=np.mean(price)
slope=np.sum((area-area_m)*(price-price_m))/np.sum((area-area_m)**2) #this is a formula for theta1
```

```python
slope
```

```python
intercept=price_m-(slope*area_m) #this is formula for theta0
intercept
```

```python
ypred=slope*area+intercept
ypred
```

```python
plt.figure()
plt.scatter(area,price)
plt.plot(area,ypred)
plt.show()
```
