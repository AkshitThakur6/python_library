```python
# selection features and target
X=df['TV']
Y=df['Sales']
```

```python
#split dataset into train and test
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
#apply Linear Regression
lr=LinearRegression()
```

```python
lr.fit(np.array(x_train).reshape(-1,1),y_train)#As fit function accept 2D features so it throing an error now here feature is 1D so we need to reshape it 
```

```python
lr.intercept_
```

```python
lr.coef_ # return slope of single feature
```

```python
#calculating y pred
y_pred=lr.predict(np.array(x_train).reshape(-1,1))
y_pred
```

```python
plt.figure()

plt.plot(x_train,y_pred)
plt.show()
```

```python
#testing 
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
y_pred
```

```python
#Evalution of model
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(mse)
print(rmse)
print(mae)
r2

```

```python
#ploynomial Regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(2) #degree of x is 2
```

```python
x=df[['TV','Radio']]
y=df['Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train_poly=pf.fit_transform(x_train)
x_test_poly=pf.transform(x_test)
```

```python
lr=LinearRegression()
lr.fit(x_train_poly,y_train)
```

```python
#testing
y_pred=lr.predict(x_test_poly)
```

```python
#evaluation of model
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
print(mse)
print(rmse)
print(mae)
print(r2)
```

```python
#residual
resid=y_test-y_pred
plt.figure()
sns.distplot(resid)
plt.show()
```
