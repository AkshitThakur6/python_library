<h3>Data Scaling/Feature Scaling...</h3>

It is process of bringging all the numeric quantity under one standard sacle<br>
it is is performed during pre precessing to handle highly varying data<br>
data scaling is not required in on linear data.<br>

2 types of scaling:
1.min max scaler<br>
standar scaler


```python
#min max scaler formula
# xnew=(x-min)/(max-min)
a=np.array([1,2,3,4,5])
b=np.array([10,20,30,60,50])
c=np.array([100,200,300,400,500])
```

```python
def minmax(arr):
    xnew=(arr-arr.min())/(arr.max()-arr.min())
```

```python
minmax(a)
a
```

```python
minmax(b)
b
```

```python
from sklearn.preprocessing import MinMaxScaler
mn=MinMaxScaler()
x.iloc[:,:]=mn.fit_transform(x.iloc[:,:])
x
```
